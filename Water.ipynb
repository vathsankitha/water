## Exploratory Data Analysis and Visualization

### Subtask:
Generate various plots (e.g., histograms, scatter plots) to understand the distribution of features, identify relationships between variables, and visualize any interesting patterns or outliers. Provide legends for all plots generated.


**Reasoning**:
The first part of the subtask is to analyze categorical feature distributions. I will start by importing the necessary libraries for plotting, then iterate through the specified categorical columns ('USO2013', 'TU', 'M') to get their unique values and counts, and finally generate bar plots for each to visualize their distributions with appropriate titles and labels.



```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_cols = ['USO2013', 'TU', 'M']

plt.figure(figsize=(18, 6))

for i, col in enumerate(categorical_cols):
    plt.subplot(1, len(categorical_cols), i + 1)

    # Get value counts and sort them for better visualization
    value_counts = df[col].value_counts().sort_index()

    sns.barplot(x=value_counts.index, y=value_counts.values, palette='viridis')

    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

plt.show()
```

**Reasoning**:
The previous code produced a `FutureWarning` from Seaborn about the `palette` argument. To fix this and follow best practices, I will modify the `sns.barplot` call to use the `hue` parameter with `value_counts.index` and set `legend=False`, as suggested by the warning message. This will ensure proper coloring and suppress the warning without changing the visual output.



```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_cols = ['USO2013', 'TU', 'M']

plt.figure(figsize=(18, 6))

for i, col in enumerate(categorical_cols):
    plt.subplot(1, len(categorical_cols), i + 1)

    # Get value counts and sort them for better visualization
    value_counts = df[col].value_counts().sort_index()

    # Fix for FutureWarning: Assign x to hue and set legend=False
    sns.barplot(x=value_counts.index, y=value_counts.values, hue=value_counts.index, palette='viridis', legend=False)

    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

plt.show()
```

**Reasoning**:
Now that the categorical feature distributions have been visualized, I will proceed to the next part of the subtask, which is to examine the numerical feature distributions. I will start by generating histograms for 'DC' and 'UL' to observe their value distributions.



```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['DC'], kde=True, color='skyblue', stat='count')
plt.title('Distribution of DC', fontsize=14)
plt.xlabel('DC', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(labels=['DC'], title='Feature')

plt.subplot(1, 2, 2)
sns.histplot(df['UL'], kde=True, color='lightcoral', stat='count')
plt.title('Distribution of UL', fontsize=14)
plt.xlabel('UL', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(labels=['UL'], title='Feature')

plt.tight_layout()
plt.show()
```
